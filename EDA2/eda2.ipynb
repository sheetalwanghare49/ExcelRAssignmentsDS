Encoding techniques

```python
#seperating categorical columns for encoding
object_list = []
for i in data_new.columns:
    if data_new[i].dtype == 'object':
        object_list.append(i)
object_list  
```

```python
print(data_new['sex'].unique())
print(data_new['income'].unique())
```

```python
data_new.head()
```

```python
#encoding the categorical into numerical
#here we encode the sex and income column using one hot encoding
data_new = pd.get_dummies(data_new, columns = ['sex','income'])
data_new
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
columnsToEncode = ['workclass','education','marital_status','occupation','relationship','race','native_country']
for x in columnsToEncode:
    data_new[x] = le.fit_transform(data_new[x])
    le_name_mapping = dict(zip(le.classes_,le.transform(le.classes_)))
    print('Feature: ',x)
    print('Mapping: ',le_name_mapping)
```

```python
data_new.head()
```

```python
'''                                                         
one hot encoding - It can lead to increased dimensionality, as a separate column is created for each category in the variable. 
                   This can make the model more complex and slow to train.It can lead to overfitting, especially if there are many 
                   categories in the variable and the sample size is relatively small.

label encoding - Assigning a unique number (starting from 0) to each class of data may lead to prioritization issues during model 
                  training of datasets. A label with a higher value may be considered to have higher priority than a label with a lower value.
'''
```

```python
#min-max scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

model = scaler.fit(data_new)
scaled_data = model.transform(data_new)
 
# print scaled features
print(scaled_data)
```

```python
#standard scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

model2 = sc.fit(data_new)
scaled_data2 = sc.transform(data_new)
```

```python
print(scaled_data2)
```

```python
'''
Choosing the appropriate scaling technique depends on the specific requirements of the algorithm, the nature of the data, and 
the desired properties of the scaled features.

* Min-Max Scaling: when you need to bring all features into a specific range.
* Standard Scaling: for algorithms that assume normally distributed data or require features to have zero mean and unit variance.
'''
```

Feature Engineering 

```python
# Create interaction term
data_new['feature1'] = data_new['capital_gain'] * data_new['capital_loss']

# Create polynomial feature (square of feature1)
data_new['feature2'] = data_new['hours_per_week'] ** 2

# Apply log transformation to feature3
data_new['log_feature3'] = np.log(data_new['fnlwgt'])
```

```python
# by applying log transformation to fnlwgt it bringing distribution closer to normal as it contains larger values
# by combining capital_gain and capital_loss we see another combining financial effect 
# by squaring hours_per_week seeing non-linear effect on outcome
```

Feature selection

```python
from sklearn.ensemble import IsolationForest
# Fit the Isolation Forest
iso_forest = IsolationForest(contamination=0.2, random_state=42)
iso_forest.fit(data_new)
outliers = iso_forest.predict(data_new)
data_new['anomaly_score'] = outliers

# Keep only non-outliers
data_clean = data_new[outliers == 1]
```

```python
data_new.head() #here we see anomaly score contain only 2 values -1 and 1(here -1=anomalies and 1=normal points)
```

```python
import ppscore as pps

# Calculate the PPS matrix
pps_matrix = pps.matrix(data_clean)

# Show the PPS matrix
print('PPS matrix:')
pps_matrix
```

```python
# Plot the PPS matrix
print('PPS Heatmap')
plt.figure(figsize=(18, 14))
sns.heatmap(pps_matrix[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore'), annot=True,cmap='cividis')
plt.show()

# Plot the correlation matrix
print('Correlation Heatmap')
plt.figure(figsize=(18, 14))
sns.heatmap(data_new.corr(), annot=True,cmap='plasma',vmax=1)
plt.show()
```
