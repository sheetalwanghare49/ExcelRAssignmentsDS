```python
#remove outlier
data_new = data[(data['Magnesium']<130)]
```

```python
#checking after removing outlier
column = data_new[["Magnesium"]]
plt.boxplot(column)
```

```python
data_new.shape
```

```python
x = data_new.values[:,:]
```

```python
x
```

```python
#scaling the data 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(x)
x = scaler.transform(x)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(x)

# Plot cumulative explained variance to determine the optimal number of components
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(7, 5))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Number of Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
# Decide on the number of components (e.g., 95% variance explained)
n_components = np.argmax(cumulative_explained_variance >= 0.95) + 1

# Step 3: Transform the original dataset into the principal components
# Apply PCA with the optimal number of components
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(x)

print(f'Number of components chosen: {n_components}')
print(f'Shape of the PCA-transformed data: {X_pca.shape}')
```

Here Applying K-means clustering

```python
from sklearn.cluster import KMeans

wsse =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, random_state=10)
    kmeans.fit(x)
    wsse.append(kmeans.inertia_)
plt.plot(range(1,11),wsse)
plt.title('The Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('WSSE')
```

```python
#here clusters are 3
kmeans = KMeans(n_clusters=3, random_state=10)
Y_pred = kmeans.fit_predict(x)
```

```python
data_new['clusters']=Y_pred
data_new.head(10)
```

```python
sns.lmplot(data=data_new , x ='Magnesium',y = 'Phenols',
           fit_reg=False, # No Regression Line
           hue="clusters",palette='Set1')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],
            s=300, c='yellow')
plt.show()
```

```python
#here we dont know important feature
```

```python
# Silhouette Score
from sklearn.metrics import silhouette_score

kmeans_silhouette = silhouette_score(x, Y_pred)
print(f'K-Means Silhouette Score: {kmeans_silhouette}')
```

```python
#Davies-Bouldin Index
from sklearn.metrics import davies_bouldin_score

db_score = davies_bouldin_score(x, Y_pred)
print(f'Davies-Bouldin Index: {db_score}')
```

Using PCA (using Sklearn)

```python
from sklearn.decomposition import PCA

# Perform PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_components = pca.fit_transform(x)

# Add PCA components to the DataFrame
data_new['PC1'] = pca_components[:, 0]
data_new['PC2'] = pca_components[:, 1]
```

```python
# Plotting the data points with clusters based on PCA components
sns.lmplot(data=data_new, x='PC1', y='PC2', 
           fit_reg=False, 
           hue="clusters", palette='Set1')


# Plotting the cluster centers (after transforming them to PCA space)
centroids_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], 
            s=300, c='yellow', label='Centroids')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering with PCA')
plt.legend()
plt.show()
```

```python
# Silhouette Score
from sklearn.metrics import silhouette_score

kmeanspca_silhouette = silhouette_score(pca_components,Y_pred)
print(f'K-Means Silhouette Score using PCA: {kmeanspca_silhouette}')
```

```python
#Davies-Bouldin Index
from sklearn.metrics import davies_bouldin_score

db_score_pca = davies_bouldin_score(pca_components,Y_pred)
print(f'Davies-Bouldin Index using PCA: {db_score_pca}')
```
