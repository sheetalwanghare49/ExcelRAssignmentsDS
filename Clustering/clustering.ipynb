```python
#removing the outlier using z_score
def remove_outliers(data, threshold=3):
    mean = data.mean()
    std = data.std()
    
    z_scores = (data - mean) / std
    
    # Filter out data points with Z-score greater than the threshold
    abs_z_scores = np.abs(z_scores)
    filtered_data = data[(abs_z_scores < threshold).all(axis=1)]
    
    return filtered_data

threshold = 3 
data_cleaned = remove_outliers(df, threshold)

# Print the number of entries before and after outlier removal
print("Number of entries before outlier removal:", len(df))
print("Number of entries after outlier removal:", len(data_cleaned))
```

K-means clustering

```python
from sklearn.cluster import KMeans

wsse =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, random_state=42)
    kmeans.fit(x)
    wsse.append(kmeans.inertia_)
plt.plot(range(1,11),wsse)
plt.title('The Elbow method')
plt.xlabel('Number of Clusters')
# Format the y-axis to avoid scientific notation
plt.ticklabel_format(style='plain', axis='y')
plt.ylabel('WSSE')
```

```python
#when clusters are three
kmeans = KMeans(n_clusters=3, random_state=42)
Y_pred = kmeans.fit_predict(x)
```

```python
data_cleaned['clusters']=Y_pred
data_cleaned.head(10)
```

```python
sns.lmplot(data=data_cleaned , x='Balance', y='Bonus_miles',fit_reg=False,
           hue="clusters",palette='Set1',height=6, aspect=1.2)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],
            s=300, c='yellow')
plt.show()
```

```python
# Silhouette Score
from sklearn.metrics import silhouette_score

kmeans_silhouette = silhouette_score(x, Y_pred)
print(f'K-Means Silhouette Score: {kmeans_silhouette}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.metrics import silhouette_score


# Linkage criteria
linkage_methods = ['single', 'complete', 'average', 'ward']
silhouette_scores = []

plt.figure(figsize=(20, 15))

for i, method in enumerate(linkage_methods):
    # Perform hierarchical clustering
    Z = linkage(x, method=method)
    
    # Cut the dendrogram to form 2 clusters
    clusters = fcluster(Z, t=2, criterion='maxclust')
    
    # Calculate the silhouette score
    silhouette_avg = silhouette_score(x, clusters)
    silhouette_scores.append(silhouette_avg)
    
    # Plot the dendrogram
    plt.subplot(2, 2, i+1)
    dendrogram(Z)
    plt.title(f'Linkage Method: {method}')

plt.tight_layout()
plt.show()

# Print silhouette scores for each linkage method
for method, score in zip(linkage_methods, silhouette_scores):
    print(f'Linkage Method: {method}, Silhouette Score: {score:.4f}')
```
